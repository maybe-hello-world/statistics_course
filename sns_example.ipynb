```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# parsing
df = pd.read_csv(r"D:\Downloads\colleges.txt", sep="\t")
df.head()
```

```python
plt.figure(figsize=(17,11))
for i, j in enumerate(['SAT', 'Acceptance', '$/Student', 'Top 10%', '%PhD', 'Grad%']):
    plt.subplot(2,3,i+1)
    sns.boxplot(x='School_Type', y=j, orient='v', data=df)
```

```python
uni = df[df['School_Type']=='Univ']
art = df[df['School_Type']=='Lib Arts']
plt.figure(figsize=(17,11))
for i, j in enumerate(['SAT', 'Acceptance', '$/Student', 'Top 10%', '%PhD', 'Grad%']):
    plt.subplot(2,3,i+1)
    sns.distplot(uni[j], label="Univ")
    sns.distplot(art[j], label="Lib Arts")
    plt.legend()
```
